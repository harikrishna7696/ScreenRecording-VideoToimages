## screen recorder

```python
# importing the required libraries
import cv2 # 1) OpenCV
import pyautogui as p #2) pyautogui for taking screen shots
import numpy as np # 3) numpy to convert images into array

size = p.size()

# saving the videp
fourcc = cv2.VideoWriter_fourcc(*"XVID")
output = cv2.VideoWriter('./imagesfromvideo/screen_recorder.mp4', fourcc, 30.0, size)

while True:
    screen_shot = p.screenshot()
    frame = np.array(screen_shot)
    output.write(frame)
    cv2.imshow('Screen Recorder', frame)
    if cv2.waitKey(1) == ord('q'):
        break
output.release()
cv2.destroyAllWindows()
```

## coverting video into frames or images

```python
import cv2

vidcap = cv2.VideoCapture('pexels.mp4')
ret,image = vidcap.read()#READ THE VIDEO
count = 0

while True:
    if ret == True:
        image = cv2.resize(image, (700, 500))
        cv2.imwrite("./imagesfromvideo/imgN%d.jpg" % count, image)     # save frame as JPEG file
#         vidcap.set(cv2.CAP_PROP_POS_MSEC,(count**100)) #used to hold speed of frane generation
        ret,image = vidcap.read()
        cv2.imshow("res",image)
        print ('Read a new frame:',count ,ret)
        count += 1
        if cv2.waitKey(10) & 0xFF == ord("q"):
            break
            cv2.destroyAllWindows()
    else:
          break

vidcap.release() 
cv2.destroyAllWindows()  
```
